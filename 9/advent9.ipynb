```python
# ─── CHALLENGE 1 ────────────────────────────────────────────────────────────────

with open('input') as file:
    lines = file.read().splitlines()

def check_number(stack, number):
    for i in range(25):
        for j in range(i+1, 25):
            if stack[i] + stack[j] == number:
                return True

numbers = list(map(lambda x:int(x), lines))
stack = numbers[:25]
n = len(lines)
i = 25

while i < n:
    if not check_number(stack, numbers[i]):
        weak_number = numbers[i]
        break
    stack.append(numbers[i])
    stack.pop(0)
    i += 1

print('result 1: ', weak_number)
```

```python
# ─── CHALLENGE 2 ────────────────────────────────────────────────────────────────

i, result2 = 0, None
while i < n:
    j, summed_numbers = 2, 0
    while summed_numbers < weak_number:
        summed_numbers = sum(numbers[i:i+j])
        if summed_numbers == weak_number:
            result2 = min(numbers[i:i+j]) + max(numbers[i:i+j])
            break
        j += 1
    if result2:
        break
    i += 1

print('result 2: ', result2)
```
