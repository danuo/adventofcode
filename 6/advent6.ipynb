```python
# ─── CHALLENGE 1 ────────────────────────────────────────────────────────────────

with open('input') as file:
    groups = file.read().split('\n\n')

num_list = []
for group in groups:
    letter_set = set()
    group_line_combined = group.replace('\n', '')
    for char in group_line_combined:
        letter_set.add(char)
    num_list.append(len(letter_set))

print('result 1:', sum(num_list))
```

```python
# ─── CHALLENGE2 ─────────────────────────────────────────────────────────────────

counter = 0
for group in groups:
    people_lines = group.split()
    
    letter_set = set()
    letter_list = []
    
    group_line_combined = group.replace('\n', '')
    for char in group_line_combined:
        letter_set.add(char)
        letter_list.append(char)
        
    n = len(people_lines)
    for letter in letter_set:
        if letter_list.count(letter) == n:
            counter += 1
    
print('result 2:', counter)
```
