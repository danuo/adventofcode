```python
# ─── CHALLENGE 1 ────────────────────────────────────────────────────────────────

from functools import reduce

with open('input') as file:
    lines = file.read().splitlines()

ticket_ids = set()
for string in lines:
    replacements = (('F', '0'), ('B', '1'), ('L', '0'), ('R', '1'))
    string_binary = reduce(lambda a, kv: a.replace(*kv), replacements, string)
    ticket_id = int(string_binary, 2) # convert '1000011110' to 542
    ticket_ids.add(ticket_id)

print('result 1:', max(ticket_ids))
```

```python
# ─── CHALLENGE 2 ────────────────────────────────────────────────────────────────

for i in range(8,2**10-8):
    if all([i-1 in ticket_ids,
            i not in ticket_ids,
            i+1 in ticket_ids]):
        print('result 2:', i)
```
