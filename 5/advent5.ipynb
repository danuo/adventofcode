```python
# ─── CHALLENGE 1 ────────────────────────────────────────────────────────────────

import numpy as np
import re

with open('input') as file:
    lines = file.read().splitlines()

def string_sanitize(string):
    return string.replace('F','0').replace('B','1').replace('L','0').replace('R','1')

def string_to_id(string):
    string = string_sanitize(string)
    string_row = string[:7]
    string_sea = string[7:]
    row = int(string_row, 2)
    seat = int(string_sea, 2)
    # return id
    return row * 8 + seat

# parse input and find highest id
highest_id = 0
for line in lines:
    id_new = string_to_id(line)
    if id_new > highest_id:
        highest_id = id_new
        
print('result 1:', highest_id)
```

```python
# ─── CHALLENGE 2 ────────────────────────────────────────────────────────────────

# collect all tickets from input as int (parsing full 10 bits)
tickets_set = set()
for line in lines:
    full_string = string_sanitize(line)
    tickets_set.add(int(full_string, 2))

# collect all neighbouring tickets as int (parsing full 10 bits), that are not
# found in the input
missing_tickets = []
for item in tickets_set:
    if item+1 not in tickets_set:
        missing_tickets.append(item+1)
    if item-1 not in tickets_set:
        missing_tickets.append(item-1)

# find missing with two neighbouring, existing tickets
for item in missing_tickets:
    if missing_tickets.count(item) > 1:
        my_ticket_int = item

# convert int of missing ticket to string of bits (base 2)
my_ticket_string = str(bin(my_ticket_int))[2:]
# convert string of bits to ticket id
my_ticket_id = string_to_id(my_ticket_string)

print('result 2:', my_ticket_id)
```
